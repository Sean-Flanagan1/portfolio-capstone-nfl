# Dropping Features in Dataset

```python
df = pd.read_csv('../../datasets/NFL Play by Play 2009-2018 (v5).csv')
```

### Drop columns with too many nulls

```python
to_drop = []
for i in df.columns:
    if df[i].isnull().sum() > 100000:
        to_drop.append(i)
```

```python
df.drop(to_drop, axis = 1, inplace = True)
```

```python
df.head()
```

### Removing WPA (win probability added) and EPA (expected points added) and plays irrelevent to a fourth down attempt

Removing WPA and EPA because these numbers change after the fourth down attempt. If WPA and EPA were not dropped, it would be telling the model the result of the play. Dropped features like qb kneel and spike, extra point and two point conversion because a team would not attempt that on fourth down. 

```python
df.drop(columns=['qb_kneel', 'qb_spike', 'total_home_rush_wpa', 'total_away_rush_wpa', 'total_home_pass_wpa', 
                 'total_away_pass_wpa', 'comp_air_wpa', 'comp_yac_wpa', 'total_home_comp_air_wpa',
                 'total_away_comp_air_wpa', 'total_home_comp_yac_wpa', 'total_away_comp_yac_wpa', 'total_home_raw_air_wpa',
                 'total_away_raw_air_wpa', 'total_home_raw_yac_wpa', 'total_away_raw_yac_wpa'], inplace = True)
```

```python
df.head()
```

```python
df.drop(columns = ['lateral_return', 'lateral_recovery', 'replay_or_challenge',
       'defensive_two_point_attempt', 'defensive_two_point_conv', 'defensive_extra_point_attempt', 
        'defensive_extra_point_conv'], inplace = True)
```

```python
df.head()
```

```python
df.drop(columns = ['assist_tackle', 'extra_point_attempt', 'extra_point_prob', 'two_point_conversion_prob'],
                  inplace = True)
```

```python
df.drop(columns = ['punt_in_endzone', 'third_down_converted', 'punt_inside_twenty', 
                   'punt_out_of_bounds', 'punt_fair_catch', 'kickoff_in_endzone', 'own_kickoff_recovery',
                  'kickoff_downed', 'two_point_attempt', 'kickoff_attempt'], inplace = True)
```
